```python
#week9

PROGRAM hw9
IMPLICIT NONE
INTEGER, PARAMETER :: ndate=60
INTEGER    :: i, j, k
INTEGER    :: mo, half, stime, etime, remainder
CHARACTER  :: year*4

INTEGER    :: yr
CHARACTER(LEN=10), DIMENSION(ndate)  :: datelist
REAL(kind=4), DIMENSION(ndate)       :: temp, running

! ----- first input (year) -----
WRITE(*, *) 'Please input the first year (2000, 2005, 2010, 2015) :'
READ(*, *)  year
if ( year .ne. '2000' .and.&
     year .ne. '2005' .and.&
     year .ne. '2010' .and.&
     year.ne.'2015') then
     print*, 'WRONG INPUT !!'
     STOP
endif

! ----- second input (number of month) -----
WRITE(*, *) 'Please input n for running mean:'
READ(*, *)  mo

remainder = mo - int(mo/2)*2
!remainder = MOD(mo,2)  ! another method with built-in function to get remainder.
IF (remainder .ne. 1) THEN
  print*, 'WRONG INPUT !!'
  stop
ENDIF

! ----- read file -----
OPEN(10, FILE='./input/t_'//year//'.txt', FORM='formatted', STATUS='old')
do i=1,ndate
READ(10, *) datelist(i), temp(i)
enddo
CLOSE(10)
! ----- calculate running mean -----
running(:) = 0.
half = (mo-1)/2
stime = half+1
etime = ndate-half
DO i=stime,etime
  ! Method 1: sum with do loop
  running(i) = 0.
  DO j=i-half, i+half
    running(i) = running(i) + temp(j)/REAL(mo)
  ENDDO

  ! Method 2: sum with built-in function
  running(i) = SUM(temp(i-half:i+half))/REAL(mo)
ENDDO

! ----- write results -----
OPEN(20, FILE='running_'//year//'.txt', FORM='formatted')
DO i=1, ndate
  if(i .lt. stime .or. i .gt. etime)then
    WRITE(20,*) datelist(i), '*'
  else
    WRITE(20, *) datelist(i), running(i)
  endif
ENDDO
CLOSE(20)


ENDPROGRAM hw9
```

```python
#work8
PROGRAM work8
IMPLICIT NONE
! This FORTRAN program reads the monthly rainfall
! data from the CWA Taipei Station in 2001–2005.

! ===============================================
! Declare the variables
INTEGER, PARAMETER:: yearnum = 5
INTEGER :: i, y, m, m_counts, max_mon, min_mon, max_year
REAL, DIMENSION(yearnum) :: ann_rain_tot
REAL :: p, ann_rain, max_rain, min_rain
! ===============================================

! Open the data file
OPEN(UNIT=10,FILE='Taipei_MonP_20012005.txt',FORM='FORMATTED',STATUS='OLD')
! Open the output file
OPEN(UNIT=20,FILE='Taipei_AnnRain_20012005.txt',FORM='FORMATTED',STATUS='UNKNOWN')
! Write the header of the ouput file
WRITE(20,*) '2001-2005 Annual Rainfall at CWA Taipei Station'
WRITE(20,*) '===================='

! Define a variable to record the month within the DO Loop
m_counts = 0

! Start reading the data with DO LOOP
DO i = 1, yearnum*12

    ! Count the month
    m_counts = m_counts + 1

    ! Read Data
    READ(10,*) y, m, p

    IF (m_counts == 1) THEN
        ann_rain = p
        max_rain = p
        max_mon  = m
        min_rain = p
        min_mon  = m
    ELSE
        ! Determine the months with the
        ! maximum and minimum rainfall
        IF (p > max_rain) THEN
            max_rain = p
            max_mon  = m
        END IF

        IF (p < min_rain) THEN
            min_rain = p
            min_mon  = m
        END IF
        ! Accumulate the annual rainfall
        ann_rain = ann_rain + p
    END IF

    ! Write results of each year into the output file
    IF (m_counts == 12) THEN
        WRITE(20,*) y
        WRITE(20,*) 'Annual Rainfall = ',ann_rain,' mm'
        WRITE(20,*) 'Maximum Monthly Rainfall = ',max_rain,' mm in ',max_mon
        WRITE(20,*) 'Minimum Monthly Rainfall = ',min_rain,' mm in ',min_mon
        WRITE(20,*) '===================='

        ! Reset and recalculate the month
        m_counts = 0
        ! Record annual rainfall into the array
        ann_rain_tot(y-2000) = ann_rain
    END IF
END DO

! Determine the year with maximum annual rainfall
max_year = 1
max_rain = ann_rain_tot(1)
DO i = 1, yearnum-1
    IF (ann_rain_tot(i+1) > max_rain) THEN
        max_year = i+1
        max_rain = ann_rain_tot(i+1)
    END IF
END DO
WRITE(20,*) 'The Year with the Maximum Annual Rainfall in 2001-2005 = ',max_year+2000

! Close files
CLOSE(10)
CLOSE(20)
END PROGRAM work8
```
